```python
# purchasing analysis (Total) 

# import required libraries
import pandas as pd
import numpy as np

# read the HeroesPymoli data
pymoli_df = pd.read_csv("/Users/rodneydavermann/Desktop/Pandas_Challenge2/Ressources/purchase_data.csv")

# number of unique items
number_unique_items = len(set(pymoli_df["Item ID"]))

# average price
average_price = round(pymoli_df["Price"].mean(), 2)

# number of purchases
number_purchases = pymoli_df["Item ID"].count()

# total revenue
total_revenue = round(pymoli_df["Price"].sum(), 2)

# create a dataframe to hold number of items, average price, number of purchases, and the total of revenue
summarize_data = pd.DataFrame({"Number of Unique Items": number_unique_items,
                               "Average Price" : [average_price],"Number of Purchases": [number_purchases],
                               "Total Revenue": [total_revenue]})
    
# format average price and total revenue to currency 
summarize_data["Average Price"] = summarize_data["Average Price"].map("${:,.2f}".format)
summarize_data["Total Revenue"] = summarize_data["Total Revenue"].map("${:,.2f}".format)
    
# display results
summarize_data
```

```python
# purchasing analysis group age

# import required libraries
import pandas as pd
import numpy as np

# read the HeroesPymoli data
pymoli_df = pd.read_csv("/Users/rodneydavermann/Desktop/Pandas_Challenge2/Ressources/purchase_data.csv")

# Create bins and bin labels for the age column
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_bins_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
    
# bin the age column
pd.cut(pymoli_df["Age"], age_bins, labels=age_bins_labels)

# append bin age group to pymoli_df
pymoli_df["Age Group"] = pd.cut(pymoli_df["Age"], age_bins, labels=age_bins_labels)
    
# group pymoli_df by Age Group
age_group_df = pymoli_df.groupby("Age Group")

# number players by age group
number_players_age_group = age_group_df.nunique()["SN"]
    
# purchase count 
purchase_count = age_group_df["Item ID"].count()
    
# average purchase price
average_purchase_price = round((age_group_df["Price"].mean()),2).map("${:,.2f}".format)

# total purchase value
total_purchase_value_age_group = round((age_group_df["Price"].sum()),2)
total_purchase_value_age_group_format = total_purchase_value_age_group.map("${:,.2f}".format)
    
# avg total purchase per person
average_total_purchase_age_group = round(total_purchase_value_age_group / number_players_age_group, 2).map("${:,.2f}".format)

# display the summary data frame
summary_data_frame = pd.DataFrame({"Purchase Count":purchase_count, 
                                   "Average Purchase Price":average_purchase_price, 
                                   "Total Purchase Value":total_purchase_value_age_group_format, 
                                   "Avg Total Purchase per Person":average_total_purchase_age_group})
summary_data_frame
```

```python
# top spenders analysis

#import required libraries
import pandas as pd
import numpy as np

# read the HeroesPymoli data
pymoli_df = pd.read_csv("/Users/rodneydavermann/Desktop/Pandas_Challenge2/Ressources/purchase_data.csv")

# group pymoli_df by SN
pymoli_df_sn = pymoli_df.groupby("SN")

# purchase count
purchase_count = pymoli_df_sn["Item ID"].count()
    
# average purchase price
average_purchase_price = pymoli_df_sn["Price"].mean()
    
# total purchase value
total_purchase_value = pymoli_df_sn["Price"].sum()
    
# top senders data frame
top_spenders_df = pd.DataFrame({"Purchase Count": purchase_count,
                                 "Average Purchase Price": average_purchase_price,
                                 "Total Purchase Value":total_purchase_value})
    
# sort the data set and display the first five records
top_senders_df_sort = top_spenders_df.sort_values(["Total Purchase Value"], ascending=False).head()

# currency format of the sorted data set
top_senders_df_sort.style.format({"Average Purchase Total":"${:,.2f}",
                                    "Average Purchase Price":"${:,.2f}", 
                                     "Total Purchase Value":"${:,.2f}"})
```

```python
# most popular items analysis

#import required libraries
import pandas as pd
import numpy as np

# read the HeroesPymoli data
pymoli_df = pd.read_csv("/Users/rodneydavermann/Desktop/Pandas_Challenge2/Ressources/purchase_data.csv")

# group pymoli df by item id and item name
pymoli_itemid_itemname_df = pymoli_df.groupby(["Item ID", "Item Name"])
    
# purchase count 
purchase_count = pymoli_itemid_itemname_df["Item ID"].count()
    
# average purchase price
average_purchase_price = pymoli_itemid_itemname_df["Price"].mean() 
    
# total purchase value
total_purchase_value = pymoli_itemid_itemname_df["Price"].sum()

# popular items data frame
popular_items_df = pd.DataFrame({"Purchase Count": purchase_count,
                                 "Average Purchase Price": average_purchase_price,
                                 "Total Purchase Value":total_purchase_value})
    
# sort the data set and display the first five records
popular_items_df_sort = popular_items_df.sort_values(["Total Purchase Value"], ascending=False).head()

# currency format of the sorted data set
popular_items_df_sort.style.format({"Average Purchase Total":"${:,.2f}",
                                    "Average Purchase Price":"${:,.2f}", 
                                     "Total Purchase Value":"${:,.2f}"})
```

```python
# most profitable items analysis

# read the HeroesPymoli data
pymoli_df = pd.read_csv("/Users/rodneydavermann/Desktop/Pandas_Challenge2/Ressources/purchase_data.csv")

# new data frame  
pymoli__items_df = pymoli_df[["Item ID", "Item Name", "Price"]]

# Group the item data by item id and item name 
pymoli__items__group_df = pymoli__items_df.groupby(["Item ID","Item Name"])

# purchase count per item 
purchase_count = pymoli__items__group_df["Price"].count()

# purchase value per item 
purchase_value = (pymoli__items__group_df["Price"].sum()) 

# calculate individual item price
individual_price = purchase_value/purchase_count

# most profitable items data frame 
profitable_items_df = pd.DataFrame({"Purchase Count": purchase_count, 
                                    "Item Price": individual_price,
                                    "Total Purchase Value":purchase_value})
    
# sort the data set and display the first five records
profitable_items_df_sort = profitable_items_df.sort_values(["Purchase Count"], ascending=False).head()

# Format with currency style
profitable_items_df_sort.style.format({"Item Price":"${:,.2f}",
                                       "Total Purchase Value":"${:,.2f}"})
```
